One major bias issue related to data availability from well-resourced versus under-resourced communities is representation bias. Well-resourced communities are more likely to have better infrastructure for data collection, such as consistent internet access, digital records, and participation in surveys or platforms that generate data. As a result, datasets can become disproportionately representative of these communities, while under-resourced communities may be underrepresented or excluded altogether.

This imbalance can lead to systematic bias in models trained on the data. For example, if a public health dataset is missing data from rural or low-income neighborhoods, any insights or policy recommendations generated may overlook the needs of those communities, potentially worsening disparities. Similarly, if crime or education datasets mostly reflect wealthier or urban areas, predictive models may inaccurately assume that the patterns in those areas apply universally.

Bias might show up in our dataset as:

Missing or incomplete entries from under-resourced areas.

Overfitting to the characteristics of well-resourced communities.

Incorrect generalizations that fail to capture the challenges or behaviors in under-resourced communities.

Ethical blind spots, such as reinforcing stereotypes or excluding marginalized groups from beneficial outcomes of data-driven decisions.

**Step 4**: Apply time series decomposition to review trends, seasonality, and residuals. New to time series analysis? Review the [Time Series Flashcard video series](https://youtube.com/playlist?list=PLNs9ZO9jGtUAqd0CNKySksPJJaOqYPMvk&feature=shared) here to learn about trends, components, and further analysis!

````python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Count the number of crashes per day, group by CRASH DATE
daily_crashes = data.groupby('CRASH DATE').size()

# Set plot style
sns.set(style="darkgrid")

# Plot the daily crashes time series
plt.figure(figsize=(15, 6))
# Use the daily_crashes Series directly instead of plt.daily_crashes
plt.plot(daily_crashes, label='Daily crashes')
plt.title('Daily Motor Vehicle Collisions in NYC')
plt.xlabel('')
# Setting the y label to a descriptive string.
plt.ylabel('Number of Crashes')
plt.legend()
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(daily_crashes, model='additive', period=365)

# Plot the decomposed components
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 12))
decomposition.trend.plot(ax=ax1)
ax1.set_title('Trend')
decomposition.seasonal.plot(ax=ax2)
# Setting the title for the seasonal plot.
ax2.set_title('Seasonality')
# Plotting residuals using ax3.
decomposition.resid.plot(ax=ax3)
ax3.set_title('Residuals')
plt.tight_layout()
plt.show()
````

The visualizations above provide valuable insights into the time series of daily motor vehicle collisions in New York City:

1. Time Series Plot: This shows the number of daily crashes over time. You might observe long-term trends, seasonal patterns, or significant outliers.

2. Decomposed Components:
  
    2.1 Trend: This graph shows the long-term trend in the data, which can indicate whether crashes are increasing, decreasing, or stable over time.

    2.2 Seasonality: This reveals any regular patterns that repeat over a specific period, such as yearly. It helps identify times of the year with higher or lower crash frequencies.

    2.3 Residuals: These are the irregular components that cannot be attributed to the trend or seasonality. They might include random or unpredictable fluctuations.

**Step 3:** Next, we will begin "Severity Mapping." We'll plot crashes on the map and code them based on severity, distinguishing between crashes that resulted in injuries and those that led to fatalities. This will give us a visual representation of where severe crashes tend to occur. </br>

You may initially want to code these incidents by using different colors, but it's important to think about your map and how accessible it is. Will a color-coded map be easy to read for everyone? Let's make a map that's inclusive for people with color blindness by also creating differently-shaped markers (squares, circles, and triangles) for crashes, injuries, and fatalities.

````python
#TODO: Continue building your heatmap
# Sample a subset of the data for visualization
sample_data_severity = data_geo.sample(n=1000, random_state=42)

# Create a base map
m_severity = folium.Map(location=[40.730610, -73.935242], zoom_start=10)

# Add crashes to the map with color coding and shape coding based on severity
for index, row in sample_data_severity.iterrows():
    if row['NUMBER OF PERSONS KILLED'] > 0:
        color = "red"  # Fatalities

        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']], # Replaced '__' and '________' with 'LATITUDE' and 'LONGITUDE' respectively
          number_of_sides=3,
          radius=5,
          gradient = False,
          color=color,
          fill=True,
          fill_color=color
        ).add_to(m_severity)


    elif row['NUMBER OF PERSONS INJURED'] > 0:
        color = "blue"  # Injuries - Replaced "_____" with "blue"
        folium.CircleMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          radius=5,
          color=color,
          fill=True,
          fill_color=color
       ).add_to(m_severity)
    else:
        color = "green"  # No injuries or fatalities - Replaced "_____" with "green"
        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          number_of_sides=4,
          radius=5,
          gradient = False,
          color=color, # Replaced "_____" with color
          fill=True,
          fill_color=color # Replaced "_____" with color
        ).add_to(m_severity)


m_severity.save("severity.html")
````

````python
#TO DO: Begin creating here!



import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import calendar
from statsmodels.tsa.seasonal import seasonal_decompose

file_path = "/content/drive/MyDrive/Motor_Vehicle_Collisions_-_Crashes_20250225.csv"

data = pd.read_csv(file_path)

# Set seaborn style
sns.set(style="darkgrid")

# --- Step 1: Ensure CRASH DATE is in datetime format
data['CRASH DATE'] = pd.to_datetime(data['CRASH DATE'])

# --- Step 2: Extract the month number from each crash date
data['MONTH'] = data['CRASH DATE'].dt.month

# --- Step 3: Group by month number and count crashes
monthly_totals = data.groupby('MONTH').size()

# --- Step 4: Convert month number to month name for labels
monthly_totals.index = [calendar.month_name[m] for m in monthly_totals.index]

# --- Step 5: Ensure months are ordered from January to December
monthly_totals = monthly_totals.reindex(calendar.month_name[1:])  # Skip the empty string at index 0

# --- Step 6: Plot the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_totals.index, y=monthly_totals.values, palette='viridis')

plt.title('Total Motor Vehicle Crashes by Month (All Years Combined)')
plt.xlabel('Month')
plt.ylabel('Number of Crashes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.savefig('monthly_crashes.png')

daily_crashes = data.groupby('CRASH DATE').size()
daily_crashes.sort_index()
year_crashes = daily_crashes['2023']

sns.lineplot(x=year_crashes.index, y=year_crashes.values, color='red')
plt.title('Total Motor Vehicle Crashes: 2023 Trend')
plt.xlabel('Day')
plt.ylabel('Number of Crashes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.savefig('daily_crashes.png')


````
